```python
import random

palavras = [
    "Amigo", "Abacaxi", "Aventura", "Animal", "Atleta", "Alegria", "Alface",
    "Abelha", "Amor", "Agulha", "Aula", "Azeite", "Assado", "Almofada",
    "Amendoim", "Aluno", "Agricultor", "Andar", "Amassar", "Amuleto", "Brinquedo",
    "Barco", "Banana", "Bebida", "Bicicleta", "Bonito", "Bolsa", "Beleza",
    "Borboleta", "Brilho", "Banho", "Brisa", "Barulho", "Bandido", "Brasa",
    "Barba", "Barraca", "Bruxa", "Baixar", "Batom", "Batida", "Beira",
    "Biblioteca", "Beethoven", "Cachorro", "Cadeira", "Caneta", "Carro",
    "Computador", "Coelho", "Chocolate", "Cachecol", "Cebola", "Cidade",
    "Camiseta", "Coruja", "Castelo", "Caneca", "Caminho", "Cacto", "Canto",
    "Cozinha", "Cachoeira", "Cogumelo", "Cenoura", "Ciclismo", "Clima", "Cama",
    "Crocodilo", "Colher", "Colina", "Caranguejo", "Corrida", "Celular", "Circo",
    "Canario", "Dente", "Desenho", "Diamante", "Duna", "Desejo", "Divertido",
    "Dose", "Dourado", "Donut", "Dieta", "Dançarino", "Dardo", "Dicionario",
    "Diamante", "Elefante", "Escola", "Estrela", "Espelho", "Esporte", "Elegante",
    "Esmalte", "Estante", "Esfera", "Energia", "Estudo", "Escada", "Escrito",
    "Entusiasmo", "Ervilha", "Engrenagem", "Enigma", "Equilibrio", "Eucalipto",
    "Escorregador", "Efeito", "Esquadro", "Esquilo", "Encontro", "Erosao",
    "Esqueleto", "Elogio", "Escrever", "Equipamento", "Escultura", "Enfermeira",
    "Elastico", "Eucaristia", "Edificio", "Eucalipto", "Esmeralda", "Embaixada",
    "Flor", "Feliz", "Fruta", "Folha", "Filme", "Festa", "Futebol", "Fantoche",
    "Fazenda", "Fotografia", "Forno", "Fonte", "Fusca", "Feriado", "Formiga",
    "Fechadura", "Folclore", "Ferrovia", "Felicidade", "Forte", "Fiacao",
    "Fundo", "Flauta", "Feitiço", "Fivela", "Futuro"
]

palavra_secreta = ""
letras_erradas = []
letras_certas = []
tentativas = 10

def main():
    print("Bem-vindo ao Jogo da Forca.")
    print("O sistema irá escolher uma palavra aleatória.\nAtenha-se às regras a seguir!")
    print(" --*-- Você terá 10 tentativas.")
    print(" --*-- Caso acerte uma letra, suas tentativas não serão alteradas.")
    print("Boa sorte!\n")
    escolher_palavra()
    jogar()

def escolher_palavra():
    global palavra_secreta
    print("Escolhendo palavra, aguarde...")
    palavra_secreta = random.choice(palavras).upper()
    print("Palavra escolhida!\nQue os jogos comecem...\n")

def jogar():
    global tentativas
    while tentativas > 0:
        mostrar_estado_atual()
        tentativa = input("\nDigite uma letra: ").upper()

        if tentativa in letras_certas or tentativa in letras_erradas:
            print("Você já tentou essa letra. Tente outra.")
            continue

        if tentativa in palavra_secreta:
            letras_certas.append(tentativa)
            print(f"Boa! A letra {tentativa} está na palavra.")
        else:
            letras_erradas.append(tentativa)
            tentativas -= 1
            print(f"Que pena! A letra {tentativa} não está na palavra. Você tem {tentativas} tentativas restantes.")

        if set(letras_certas) == set(palavra_secreta):
            print("\nParabéns! Você adivinhou a palavra!")
            print(f"\nA palavra era {palavra_secreta}.")
            break
    else:
        print(f"\nVocê perdeu! A palavra era {palavra_secreta}.")

def mostrar_estado_atual():
    estado_atual = [letra if letra in letras_certas else "_" for letra in palavra_secreta]
    print("Palavra: ", " ".join(estado_atual))
    print("Letras erradas: ", " ".join(letras_erradas))

main()
```
